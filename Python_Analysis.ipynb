```python
GENDER
```

```python
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df)
```

```python
#for labels 
ax=sns.countplot(x='Gender',data=df) 
for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False) 
ay=sns.barplot(x='Gender',y='Amount',data=sales_gen) 
for bars in ay.containers: 
    ay.bar_label(bars)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

AGE

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender') 
for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False) 
ay=sns.barplot(x='Age Group',y='Amount',data=sales_age) 
for bars in ay.containers: 
    ay.bar_label(bars)
```

From above Graph we can see that most of the buyers are of age group between 26-35 yrs female

STATE

```python
df.columns
```

```python
#total no. of orders from top 10 states 
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10) 

ay=sns.barplot(x='State',y='Amount',data=sales_state) 
for bars in ay.containers: 
    ay.bar_label(bars)
```

```python
#for size
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10) 
sns.set(rc={'figure.figsize':(15,5)})
ay=sns.barplot(x='State',y='Amount',data=sales_state) 
for bars in ay.containers: 
    ay.bar_label(bars)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10) 
sns.set(rc={'figure.figsize':(15,5)})
ay=sns.barplot(x='State',y='Orders',data=sales_state)  
for bars in ay.containers: 
    ay.bar_label(bars)
```

From above graphs we can see that most of the orders and total sales/amount are from Uttar pradesh

```python
ax=sns.countplot(data=df,x='Marital_Status') 
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers: 
    ax.bar_label(bars)
```

OCCUPATION

```python
ax=sns.countplot(data=df,x='Occupation') 
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,5)})
ay=sns.barplot(x='Occupation',y='Amount',data=sales_state) 
for bars in ay.containers: 
    ay.bar_label(bars)
```

From above graphs we can see that most of the buyers are working in IT Sector

PRODUCT CATEGORY

```python
ax=sns.countplot(data=df,x='Product_Category') 
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
ay=sns.barplot(x='Product_Category',y='Amount',data=sales_state) 
for bars in ay.containers: 
    ay.bar_label(bars)
```

From above graphs we can see that most of the sold products are food

```python
sales_state=df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
ay=sns.barplot(x='Product_ID',y='Amount',data=sales_state) 
for bars in ay.containers: 
    ay.bar_label(bars)
```
